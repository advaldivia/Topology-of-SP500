```python
#pick window size and make windows, done in halves because of hardware constraints
w = 50
windows = [(2*(1-(ratios.iloc[(i):((i)+w),:].corr()))).transform(np.sqrt).dropna(axis='columns',how='all').dropna(axis='rows',how = 'all').to_numpy() 
                for i in range(ratios.shape[0] - (w+1))]
first_half = windows[0:3000]
second_half =windows[3000:]

#make persistence diagrams
def dist_rip(data):
    return ripser.ripser(data, distance_matrix = True, maxdim = 2)
persist_by_window_1 = [dist_rip(item) for item in first_half ]
persist_by_window_2 = [dist_rip(item) for item in second_half]

#save results
np.save('dgms_1.npy',persist_by_window_1, allow_pickle = True)
np.save('dgms_2.npy',persist_by_window_2, allow_pickle = True)
```

```python
#Put into one array
persist_by_window = np.concatenate((persist_by_window_1,persist_by_window_2))
```

```python
## Given a persistence diagram we want to create a persistence landscape.

## First construct function.  Inputs are birth death pairs in nX2 array , d (depth), x (value to evaluate the function at)
from numpy.linalg import norm 
def landscape(dg, d, x):
    L = []
    for i in range(dg.shape[0]):
        if ((x>dg[i,0]) &(x<dg[i,1])):
            L.append(dg[i,:])
    H = []
    for i in range(len(L)):
        if x<=(sum(L[i])/2):
            H.append(x-L[i][0])
        else:
            H.append(L[i][1]-x)
    
    H.sort(reverse = True)
    
    if len(H)>d:
        return(H[d])
    else:
        return(0)



##Numerical integration for Lp Norm depends on p, and number of points sampled in range
def landscaper_norm(n,p,dg):
    low = min([dg[i,0] for i in range(dg.shape[0])])
    high = max([dg[i,1] for i in range(dg.shape[0])])
    delta = (high-low)/n
    
    ints = [(1/n)*sum([landscape(dg,d,low+(i/n)) for i in range(n)]) for d in range(dg.shape[0])]
    
    value = norm(ints,p)
    
    return value
```

```python
#compute landscape persistence norms in dimensions 1 and 2
p_norm_0 = [landscaper_norm(100,1,persist_by_window[i]['dgms'][0]) for i in range(len(persist_by_window))]
np.save('persistence_norms_00_to_23_dim0.npy',p_norm_1, allow_pickle = True)
p_norm_1 = [landscaper_norm(100,1,persist_by_window[i]['dgms'][1]) for i in range(len(persist_by_window))]
np.save('persistence_norms_00_to_23.npy',p_norm_1, allow_pickle = True)
p_norm_2 = [landscaper_norm(100,1,persist_by_window[i]['dgms'][2]) for i in range(len(persist_by_window))]
np.save('persistence_norms_00_to_23_dim2.npy',p_norm_2, allow_pickle = True)
```

```python
#measure of Euler characteristic
euler_norm = [p_norm_2[i]/p_norm_1[i] for i in range(len(p_norm_1))]
```
